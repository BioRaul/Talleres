**COMANDOS PARA GRAFICAR EL PLÁSMIDO**

```python
caracts=[GraphicFeature(start=10, end=140, strand=+1,color="orange", label="Promotor1"),
        GraphicFeature(start=150, end=590, strand=+1,color="blue", label="Gen1"),
        GraphicFeature(start=400, end=912, strand=-1,color="red", label="Gen2"),
        GraphicFeature(start=910, end=1035, strand=-1,color="orange", label="Promotor2")]

graphic=GraphicRecord(sequence_length=1000, features=caracts)

graphic.plot()
```

```python
graphic1=CircularGraphicRecord(sequence_length=1000, features=caracts)
graphic1.plot()
```
